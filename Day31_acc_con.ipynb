```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
x,y = load_breast_cancer(return_X_y=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
y_pred1 = lr.predict(x_test)
accuracy_score(y_test,y_pred1)
```

```python
dc = DecisionTreeClassifier()
```

```python
dc.fit(x_train,y_train)
y_pred2 = dc.predict(x_test)
accuracy_score(y_test,y_pred2)
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
confusion_matrix(y_test,y_pred2)
```

```python
print("for Logistic Regression Model")
print("-"*40)
cfd = pd.DataFrame(confusion_matrix(y_test,y_pred1), columns=["Predicted Negative","Predicted Positive"])
print(cfd)
# print(f"Confusion Matrix: \n{confusion_matrix(y_test,y_pred1)}")
print(f"Accuracy Score: {accuracy_score(y_test,y_pred1)}")
print(f"Recall score: {recall_score(y_test,y_pred1)}")
print(f"Precision Score: {precision_score(y_test,y_pred1)}")
print(f"F1 Score: {f1_score(y_test,y_pred1)}")
```

```python
print("for Decision Tree Regression Model")
print("-"*40)
cfd = pd.DataFrame(confusion_matrix(y_test,y_pred2), columns=["Predicted Negative","Predicted Positive"])
print(cfd)
# print(f"Confusion Matrix: \n{confusion_matrix(y_test,y_pred1)}")
print(f"Accuracy Score: {accuracy_score(y_test,y_pred2)}")
print(f"Recall score: {recall_score(y_test,y_pred2)}")
print(f"Precision Score: {precision_score(y_test,y_pred2)}")
print(f"F1 Score: {f1_score(y_test,y_pred2)}")
```
