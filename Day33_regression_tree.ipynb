```python
import numpy as np
import pandas as pd
from pandas_datareader import data
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import GridSearchCV
```

```python
cald = fetch_california_housing()
df = pd.DataFrame(cald.data, columns=cald.feature_names)
df['target'] = cald.target
```

```python
df
```

```python
x = df.drop(['target'], axis=1)
y = df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
regt = DecisionTreeRegressor(criterion = 'friedman_mse',max_depth=5)
regt.fit(x_train, y_train)
```

```python
y_pred = regt.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

Hyperparameter Tuning

```python
param_grid = {
    'max_depth': [2,4,8,10,None],
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'max_features': [4,6,8],
    'min_samples_split': [2,4,8,10]
}
```

```python
reg = GridSearchCV(DecisionTreeRegressor(), param_grid, cv=5)
reg.fit(x_train, y_train)
```

```python
reg.best_score_
```

```python
print(reg.best_params_)
```

Feature importance

```python
for importance, name in sorted(zip(reg.best_estimator_.feature_importances_, x_train.columns), reverse=True):
    print(name, importance)
```
