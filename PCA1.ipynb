```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(23)

mu_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)

df = pd.DataFrame(class1_sample,columns=['x1','x2','x3'])
df['target'] = 1


mu_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20)

df1 = pd.DataFrame(class2_sample,columns=['x1','x2','x3'])
df1['target'] = 0

df = pd.concat([df, df1],ignore_index=True)
df.head()
```

Applying scaler to the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df.iloc[ : , :3] = scaler.fit_transform(df.iloc[: , :3])
```

```python
#finding covarience matrix
cov_matrix = np.cov([df['x1'], df['x2'], df['x3']])
cov_matrix
```

```python
#eigen vectos and eigen values

eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)
print(eigen_values)
print(eigen_vectors)
```

```python
pc = eigen_vectors[0:2]
pc
```

```python
transformed_df = np.dot(df.iloc[ : , :3], pc.T)
transformed_df_new = pd.DataFrame(transformed_df, columns=['PC1', 'PC2'])
transformed_df_new.head()
```

```python
transformed_df_new['target'] = df['target']
transformed_df_new.head()
```
