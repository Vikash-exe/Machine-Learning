```python
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
```

```python
x,y = make_classification(n_classes=2, n_features=2, n_samples=100, n_informative=1, n_clusters_per_class=1, random_state=41, hypercube=False, n_redundant=0, class_sep=20)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0], x[:,1], c=y,cmap='winter', s=100)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(penalty = 'l2', solver= 'sag')
lor.fit(x,y)
```

```python
print(lor.coef_)
print(lor.intercept_)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
c = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input+ c
```

logistic gradient descent

```python


def gd(x,y):

  x = np.insert(x,0,1,axis=1)
  weights = np.zeros(x.shape[1])
  lr = 0.5

  for i in range(5000):
    y_hat = sigmoid(np.dot(x,weights))
    weights = weights + lr * ((np.dot((y-y_hat),x))/x.shape[0])

  return weights[1:],weights[0]
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
coef_, intercept_ = gd(x,y)
```

```python
m1 = -(coef_[0]/coef_[1])
c1 = -(intercept_/coef_[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m1*x_input + c1
```

```python
plt.figure(figsize=(12,6))
plt.plot(x_input,y_input, label='sklearn', color = 'blue')
plt.plot(x_input1,y_input1, label='gd', color='red')
plt.scatter(x[:,0], x[:,1], c=y,cmap='winter', s=100)
plt.ylim(-4, 4) # Adjust y-axis limits to better view the lines
plt.xlim(-4, 4) # Adjust x-axis limits to better view the lines
plt.legend()
plt.show()
```
