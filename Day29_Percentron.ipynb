```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import Lasso, LinearRegression, Ridge, ElasticNet
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_classes=2,
                          n_clusters_per_class=1, hypercube=False, class_sep=10, random_state=41)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x[:,0], x[:,1],c=y, cmap='rainbow')
```

```python
def step(z):
  return 1 if z>0 else 0

def perceptron(x,y):
  x = np.insert(x,0,1,axis=1)
  weights = np.ones(x.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,x.shape[0])
    y_hat = step(np.dot(x[j],weights))
    weights = weights + lr * (y[j]-y_hat) * x[j]

  return weights[0],weights[1:]

```

```python
intercept_,coef_ = perceptron(x,y)
```

```python
print(coef_,intercept_)
```

```python
m = -(coef_[0]/coef_[1])
c = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input+ c
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x[:,0], x[:,1],c=y, cmap='rainbow')
plt.plot(x_input,y_input,color='blue')
plt.ylim(-2,2)
```

# Sigmoid funcion

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def step(z):
  return 1 if z>0 else 0

def perceptron(x,y):
  x = np.insert(x,0,1,axis=1)
  weights = np.ones(x.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,x.shape[0])
    y_hat = sigmoid(np.dot(x[j],weights))
    weights = weights + lr * (y[j]-y_hat) * x[j]

  return weights[0],weights[1:]
```
