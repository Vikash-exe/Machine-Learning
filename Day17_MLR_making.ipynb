# Linear regression class making

```python
class mylr:
  def __init__(self):
    self.coef_ = None        #betas for all features
    self.intercept_ = None   # B0 i.e. the scaler value of y

  def fit(self, x_train, y_train):
    x_train = np.insert(x_train,0,1,axis=1) #inserting column with value 1 at index[0]

    #coeff calculation
    betas = np.linalg.inv(np.dot(x_train.T,x_train)).dot(x_train.T).dot(y_train)  #np.linalg.inv for inverse and run the expression for whole x_train
    #print(betas)
    self.intercept_ = betas[0]
    self.coef_ = betas[1:]

  def predict(self, x_test):
    y_pred = np.dot(x_test, self.coef_) + self.intercept_
    return y_pred
```

```python
lrn = mylr()
```

```python
lrn.fit(x_train,y_train)
```

```python
print(lrn.intercept_)
print(lrn.coef_)
```

```python
y_pred = lrn.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```
